# Cleaning Data

Only two columns contain null values. Only the 'days_occupied_in_2018' column is less than the threshold, so it can be removed. Then tranformed the 'reviews_per_month' column's data type from float to int to turn it into a whole number, as it makes more sense to have a whole number of reviews in a month.

```python
#making copy of df
copy = pd.read_csv('/content/drive/MyDrive/School/grad/2024/Fall 24/Data Visualization for Business/Quiz 2/san_francisco.csv')

#figure out how manu Null values
print(df.isna().sum())
#create threshold
threshold = len(df) * 0.5
print("threshold: " +str(threshold))

#compare threshold to columns to decide to drop
cols_to_drop = df.columns[df.isna().sum() >= threshold]
print("cols_to_drop: " + str(cols_to_drop))

#drop days_occupied_in_2018
df = df.drop('days_occupied_in_2018', axis=1)
```

```python
df.info()
```

```python
#find unique values
print(df.nunique())
```

```python
#round reviews_per_month into whole number
df['reviews_per_month'].round()

df['reviews_per_month'].fillna(0,inplace=True)

#change reviews_per_month into int datatype
df['reviews_per_month'] = df['reviews_per_month'].astype(int)
```

```python
#convert neighborhood to string dtype
df['neighbourhood'] = df['neighbourhood'].astype(str)
```
